```python
# Equality: asking if values are the same
x:  int = 10
y: int = 5

equal: bool = x == y
print(f"{x} == {y}: {equal}")
```

```python
# Equality: asking if values are the same
x != 10
y != 5

not_equal: bool = x != y
print(f"{x} != {y}: {not-equal}")
```

```python
# Greater than and less than
x = 10
y = 5
greater = x > y
print(f"{x} > {y}: {greater}")
```

```python
# Greater than and less than
x = 10
y = 5
less_than = x < y
print(f"{x} < {y}: {less_than}")
```
